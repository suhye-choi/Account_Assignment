```python
class Square:
    def __init__(self):
        self.square = int(input('ë„“ì´ë¥¼ êµ¬í•˜ê³  ì‹¶ì€ ì‚¬ê°í˜•ì˜ ìˆ«ìë¥¼ ì¨ì£¼ì„¸ìš”.\n 1.ì§ì‚¬ê°í˜• 2.í‰í–‰ì‚¬ë³€í˜• 3.ì‚¬ë‹¤ë¦¬ê¼´ \n >>> '))

        if self.square == 1:
            print('ì§ì‚¬ê°í˜• í•¨ìˆ˜ëŠ” rect()ì…ë‹ˆë‹¤.')

        elif self.square == 2:
            print('í‰í–‰ì‚¬ë³€í˜• í•¨ìˆ˜ëŠ” par()ì…ë‹ˆë‹¤.')
        
        elif self.square == 3:
            print('ì‚¬ë‹¤ë¦¬ê¼´ í•¨ìˆ˜ëŠ” trape()ì…ë‹ˆë‹¤.')
        
        else:
            print('1, 2, 3 ì¤‘ì—ì„œ ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”')


    def rect(self):
        width, vertical = map(int, input('ê°€ë¡œ, ì„¸ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ê°€ë¡œ,ì„¸ë¡œ\n >>> ').split(','))
        area = width * vertical
        result = f'ì§ì‚¬ê°í˜•ì˜ ë„“ì´ëŠ” : {area}'
        return result

    def par(self):
        width, height = map(int, input('ê°€ë¡œ, ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ê°€ë¡œ,ë†’ì´\n >>> ').split(','))
        area = width * height
        result = f'í‰í–‰ì‚¬ë³€í˜•ì˜ ë„“ì´ëŠ” : {area}'
        return result

    def trape(self):
        width1, width2, height = map(int, input('ìœ—ë©´, ì•„ë«ë©´, ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ìœ—ë©´,ì•„ë«ë©´,ë†’ì´\n >>> ').split(','))
        area = (width1 + width2) * height / 2
        result = f'ì‚¬ë‹¤ë¦¬ê¼´ì˜ ë„“ì´ëŠ” : {area}'
        return result


# ì‹¤í–‰ ì˜ˆì‹œ
a = Square()

# í‰í–‰ì‚¬ë³€í˜•ì˜ ë„“ì´
print(a.par())  # 3,2 ì…ë ¥ â†’ 6

# ì‚¬ë‹¤ë¦¬ê¼´ì˜ ë„“ì´
print(a.trape())  # 3,2,1 ì…ë ¥ â†’ 2.5
```

```python
class Kiosk:    
    def __init__(self):
        self.menu = ['ì•„ë©”ë¦¬ì¹´ë…¸', 'ë¼ë–¼', 'ì½œë¼']
        self.price = [2000, 3000, 1500]
        self.order_menu = []     
        self.order_price = []    
        self.price_sum = 0       
        self.pay = None

    def menu_print(self):
        print("\nğŸ“‹ ë©”ë‰´íŒ")
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], 'ì›')

    def menu_select(self):
        while True:
            try:
                n = int(input("\nì¶”ê°€ ì£¼ë¬¸í•  ìŒë£Œ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì§€ë¶ˆì€ 0): "))
            except ValueError:
                print("ìˆ«ìë§Œ ì…ë ¥í•˜ì„¸ìš”.")
                continue

            if n == 0:
                if not self.order_menu:
                    print("ì£¼ë¬¸ì´ ë¹„ì–´ ìˆìŠµë‹ˆë‹¤. ìµœì†Œ 1ì”ì€ ì„ íƒí•˜ì„¸ìš”.")
                    continue
                break

            if 1 <= n <= len(self.menu):
                temp_choice = 0
                while temp_choice not in (1, 2):
                    try:
                        temp_choice = int(input("HOTì€ 1, ICEëŠ” 2ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
                    except ValueError:
                        print("ìˆ«ìë§Œ ì…ë ¥í•˜ì„¸ìš”.")
                        continue

                    if temp_choice not in (1, 2):
                        print("ìŒë£Œì˜ ì˜¨ë„ë¥¼ ì„ íƒí•´ì£¼ì„¸ìš”")

                temp_str = "HOT" if temp_choice == 1 else "ICE"
                drink_name = f"{temp_str} {self.menu[n-1]}"
                drink_price = self.price[n-1]

                self.order_menu.append(drink_name)
                self.order_price.append(drink_price)
                self.price_sum += drink_price

                print(f"ì¶”ê°€ ì£¼ë¬¸: {drink_name} ({drink_price}ì›)")
                print(f"í˜„ì¬ í•©ê³„: {self.price_sum}ì›")
            else:
                print("ì—†ëŠ” ë©”ë‰´ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•˜ì„¸ìš”.")

    def pay_select(self):
        pay_choice = 0
        while pay_choice not in (1, 2):
            try:
                pay_choice = int(input("\nê²°ì œ ë°©ë²•ì„ ì„ íƒí•˜ì„¸ìš” (1:í˜„ê¸ˆ / 2:ì¹´ë“œ): "))
            except ValueError:
                print("ìˆ«ìë§Œ ì…ë ¥í•˜ì„¸ìš”.")
                continue

            if pay_choice == 1:
                self.pay = "cash"
                print("ì§ì›ì„ í˜¸ì¶œí•˜ê² ìŠµë‹ˆë‹¤.")
            elif pay_choice == 2:
                self.pay = "card"
                print("ICì¹© ë°©í–¥ì— ë§ê²Œ ì¹´ë“œë¥¼ ê½‚ì•„ì£¼ì„¸ìš”.")
            else:
                print("1 ë˜ëŠ” 2ë§Œ ì…ë ¥í•˜ì„¸ìš”.")

    def print_receipt(self):
        print('\nâŸ' + '-' * 30 + 'âŸ')
        print('|{:^30}|'.format("ì£¼ë¬¸ ë‚´ì—­"))
        if not self.order_menu:
            print('| {:<28}|'.format("ì¥ë°”êµ¬ë‹ˆê°€ ë¹„ì—ˆìŠµë‹ˆë‹¤."))
        else:
            for name, price in zip(self.order_menu, self.order_price):
                line = f"{name} : {price}ì›"
                print('| {:<28}|'.format(line))
        print('|' + ' ' * 30 + '|')
        print('| {:<28}|'.format(f"ì´ ê¸ˆì•¡ : {self.price_sum}ì›"))
        print('| {:<28}|'.format(f"ê²°ì œ ë°©ë²• : {self.pay}"))
        print('âŸ' + '-' * 30 + 'âŸ')


# ì‹¤í–‰ ìˆœì„œ ì˜ˆì‹œ
a = Kiosk()
a.menu_print()
a.menu_select()   # ì—¬ëŸ¬ ì” ë‹´ê¸° (0 ì…ë ¥ ì‹œ ì¢…ë£Œ)
a.pay_select()    # ê²°ì œ ì„ íƒ
a.print_receipt() # ì˜ìˆ˜ì¦ ì¶œë ¥
```
